# Logistic Regression Hyphothesis

# $h_{\theta}(x)=g(\theta^T x)$

.
# Where g is sigmoind function

#  $g(z)=\frac{1}{1+e^{-z}}$

```python
def sigmoid(z):
    return  1 / (1 + np.exp(-z))
```

# Neural Network Cost Function
# $J(\theta)=\frac{1}{m}{\sum\limits_{K=1}^{K}}{\sum\limits_{i=1}^{m}}(-y_K^{(i)}log(h_{\theta}(x^{(i)})_K)-(1-y_K^{(i)})log(1-h_{\theta}(x^{(i)}))_K)+\\
\frac{\lambda}{2m} \bigg[\sum\limits_{K=1}^{400} \sum\limits_{i=1}^{25}(\theta^{(1)}_{j,K})^2 
+ \sum\limits_{K=1}^{25} \sum\limits_{i=1}^{10}(\theta^{(1)}_{j,K})^2 \bigg]$

```python
# Neural Network (NN) cost function
def nnCostFunction(nn_params,input_layer_size,hidden_layer_size,K,X, y, lambda_):
    # K is number of units in output layer
    
    
    nn_params = np.concatenate([Theta1.ravel(), Theta2.ravel()])
    
    
    m=y.size
    J=0
    # Part 1: Forward propagation
    #
    # Input layer(layer1)
    # Let, a1 = X, and add a column of bias units to a1(as the first column)
    a1 = np.concatenate([np.ones((m, 1)), X], axis=1)  # 5000 x 401
         
    # add ones to X data matrix
    X = np.hstack((np.ones((m, 1)), X))
            
    # hidden layer
    z2 = np.dot(X, Theta1.T) #(5000 x 401)*(401 x 25) -> 5000 x 25
    a2 = sigmoid(z2)
    
    #add bias unit to a2
    a2 = np.hstack((np.ones((a2.shape[0],1)), a2))

    # output layer
    z3 = np.dot(a2, Theta2.T) # (5000 x 26)*(26 x 10) -> 5000 x 10
    a3 = sigmoid(z3)  # 5000 x 10
    
    # Encode the labels as vectors containing only values 0 or 1
    # np.eye returns a 2-D array with ones on the diagonal and zeros elsewhere.
    # reshape(-1): Flattening array means converting a multidimensional array into a 1D array. We can use reshape(-1) to do this.
    y_matrix = np.eye(num_labels)[y.reshape(-1)]
    
    J = (-1/m) * (np.sum(y_matrix * np.log(a3) + (1-y_matrix) * np.log(1-a3)))
            

    #regularization term excluding the Theta columns for the bias units
    Reg = (lambda_/(2*m)) * (np.sum(np.square(Theta1[:, 1:])) + np.sum(np.square(Theta2[:, 1:]))) 
    J = J + Reg
    
    return J
```

```python
lambda_ = 0
J= nnCostFunction(nn_params, input_layer_size, hidden_layer_size,
                   num_labels, X, y, lambda_)
J
```

```python
lambda_ = 1
J= nnCostFunction(nn_params, input_layer_size, hidden_layer_size,
                   num_labels, X, y, lambda_)
J
```
