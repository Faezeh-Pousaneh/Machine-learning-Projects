# Normalize features, define covariance matrix 'Covariance', define PCA

### SVD give the eigenvectors  and eigenvalues of matrix of covariance. For PCA we choose the biger eigenvector which comes from bigger eigenvalues. The lenght of eigenvector is eigenvalue. 

```python
# in the image below $\sum$ indicates covariance
from IPython import display
display.Image("./Image.png", width=700)
```

```python
# X is a 2D and we want to reduce to 1D
X = mydata['X']
```

```python
def featureNormalize(X):
    
    mu = np.mean(X, axis=0)    
    X_norm = (X - mu)/np.std(X, axis=0)
    
    return X_norm, mu


def pca(X):
    
    m, n = X.shape
    # compute covariance matrix
    Covariance = np.dot(X.T, X)/m
    # SVD, eigenvectors of cocariance
    # u is eigenvectors of original data and s is matrix of eigenvalues. the 
    u, s, v = np.linalg.svd(Covariance)
    
    return u,s

def drawLine(p1, p2, *args, **kwargs):
    
    plt.plot([p1[0], p2[0]], [p1[1], p2[1]], *args, **kwargs)
```

### Example: what do u,s,vh mean

## Plot eigenvector of the data

```python
# feature normalize
X_norm, mu = featureNormalize(X)

# perform PCA
U, S = pca(X_norm)
```

```python
plt.figure(figsize=(5,5))
plt.plot(X[:,0], X[:,1], 'bo')
# u is eigenvectors and s is matrix of eigenvalues
#first eigenvector and second. the larger is for S[0]=1.735 the bigger eighenvalue whitch is our pca.
drawLine(mu, mu+1.5*S[0]*U[:,0], lw=2, color='k')
drawLine(mu, mu+1.5*S[1]*U[:,1], lw=2, color='k')
plt.xlim(0.5, 7)
```

# Projecting Data into PCA component

```python
#define the projection function to pca
def projectData(X, U, K):
    
    return np.dot(X, U[:,:K])
#going back from pca to approx data
def recoverData(Z, U, K):
    
    return np.dot(Z, U[:,:K].T)
```

```python
K = 1
Z = projectData(X_norm, U, 1)
print(Z.shape, Z[0])
```

### Reconstructing approximate data

```python
X_recover = recoverData(Z, U, K)
print(X_recover[0,0], X_recover[0,1])
```

```python
plt.figure(figsize=(5,5))
plt.plot(X_norm[:,0], X_norm[:,1], 'bo')
plt.plot(X_recover[:,0], X_recover[:,1], 'ro')
for ii in range(X_norm.shape[0]):
    drawLine(X_norm[ii,:], X_recover[ii,:], '--k')
plt.xlim(-4,3)
plt.ylim(-4,3)
```
